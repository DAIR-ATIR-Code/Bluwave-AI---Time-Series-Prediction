```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from math import sqrt
```

### Load data

```python
target = 'Wind Spd (km/h)'
clean = pd.read_csv(str(Path("../data/interim/clean.csv")), parse_dates=True, 
                    infer_datetime_format=True, index_col=0, low_memory=False)
predict = pd.read_csv(str(Path("../models/predictions.csv")), parse_dates=True, 
                      infer_datetime_format=True, index_col=0, header=None, names=[target])
training_history = pd.read_csv(str(Path("../models/training_history.csv")), index_col=0)
```

```python
train_end = '2016-12'
val_start = '2017-01'
val_end = '2017-12'
test_start = '2018-01'
    
train = clean.loc[:train_end]
validate = clean.loc[val_start:val_end]
test = clean.loc[test_start:]
    
train_y = train.pop(target)
validate_y = validate.pop(target)
test_y = test.pop(target)   
    
persist_y = pd.Series(test_y.shift(1), index=test.index)
predict_y = predict[target]

predict_se = (test_y - predict_y)**2
persist_se = (test_y - persist_y)**2
predict_rmse = sqrt(predict_se.mean())
persist_rmse = sqrt(persist_se.mean())
```

### Inspect performance of model training

```python
# Plot objective function over # of epochs to examine the performance of model training
fig, ax = plt.subplots(1,1,figsize=(15,5))
ax.semilogy(training_history['train_loss'].iloc[100:], label='training data')
ax.semilogy(training_history['validate_loss'].iloc[100:], label='validation data')
ax.set_ylabel('mse')
ax.set_xlabel('# of epochs')
plt.legend()
plt.show()
```

```python
# Analyze learning curves:
# - Both error curves are smooth, so we chose a good learning rate for our optimizer
# - Both error curves plateau, so we trained for sufficiently many epochs
# - Training error < validation error to a reasonable degree
```

### Inspect performance of predictor

```python
# Plot all data
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.plot(train_y, label='training data')   
ax.plot(validate_y, label='validation data')
ax.plot(test_y, label='testing data (actual)')
ax.plot(predict_y, label='testing data (prediction)')
ax.legend(bbox_to_anchor=(1.05,0.5))
plt.show()
```

```python
# Compare predictions to actual test data
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.plot(test_y, label='testing data (actual)')
ax.plot(predict_y, label='testing data (prediction)')
ax.legend(bbox_to_anchor=(1.05,0.5))
plt.show()
```

```python
# Compare predictions to actual test data on a smaller scale
fig, ax = plt.subplots(1,1,figsize=(15,5))
ax.plot(test_y[:75], label='testing data (actual)')
ax.plot(predict_y[:75], label='testing data (prediction)')
#ax.plot(persist_y[:75], label='testing data (persistence)')
plt.legend()
plt.show()
```

```python
# Compare error of prediction versus error of persistence
fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.bar([0,1], height=[predict_rmse, persist_rmse])
plt.show()
```
